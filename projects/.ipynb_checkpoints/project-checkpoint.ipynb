```python
def calculate_loyalty_ranking(point):
    point = int(point)
    if point in range(1, 1000):
        return "STANDARD"
    if point in range(1000, 2000):
        return "SILVER"
    if point in range(2000, 5000):
        return "GOLD"
    else:
        return "DIAMOND"
```

```python
def calculate_point(index_transaction, row_transaction):
    for index, row in loyalty_points_df["Point Mechanism"].items():
            if(loyalty_points_df.at[index, "Service Group"].lower() == row_transaction["Service Group"].lower()):
                    ## Point Mechanism Handle
                    point_mechanism = row[:row.index("VND")].split("points/")
                    point = int(point_mechanism[0])
                    unit_gmv = int(point_mechanism[1] )
                    
                    ## Calculate point
                    calculate_point = (row_transaction["GMV"] * point) / unit_gmv
                    maximum_point = int(loyalty_points_df.at[index, "Maximum Point Per Trans"].split(" points")[0])
                    if (calculate_point > maximum_point):
                        calculate_point = maximum_point
                    transaction_df.at[index_transaction, "Calculated Point"] = calculate_point 
```

```python
for index, row in transaction_df.iterrows():
    calculate_point(index, row)
transaction_df.sample(5)    
```

```python
saved_uid_dict = {}
for index, row in transaction_df.iterrows():
    if row["User_id"] not in saved_uid_dict.keys():
        saved_uid_dict[row["User_id"]] = {}
        saved_uid_dict[row["User_id"]]["Calculated Point"] = []
        saved_uid_dict[row["User_id"]]["DATE"] = []
        saved_uid_dict[row["User_id"]]["Order_id"] = []
    else:
        saved_uid_dict[row["User_id"]]["Calculated Point"].append(row["Calculated Point"])  
        saved_uid_dict[row["User_id"]]["DATE"].append(row["DATE"])         
        saved_uid_dict[row["User_id"]]["Order_id"].append(row["Order_id"])         
```

```python
def get_accumulated_point(index, row):
    result = 0
    uid = saved_uid_dict.get(row["User_id"])
    current_day = row["DATE"]
    if(len(uid.get("Calculated Point")) > 0):
       for i in range(len(uid.get("Calculated Point"))):
            if ((current_day - uid.get("DATE")[i]).days <= 30) & \
                ((current_day - uid.get("DATE")[i]).days >= 0) & \
                (uid.get("Order_id")[i] != row["Order_id"]):
                result += uid.get("Calculated Point")[i]
            else: 
                return result     
    return result        
```

```python
for index, row in transaction_df.iterrows():
    transaction_df.at[index, "Accumulated Point"] = get_accumulated_point(index, row)
```

```python
for index, row in transaction_df.iterrows():
    transaction_df.at[index, "Rank_name"] = calculate_loyalty_ranking(row["Calculated Point"] + row["Accumulated Point"])   
transaction_df.sample(5)    
```

```python
lst_week = pd.date_range(transaction_df["DATE"].min(), transaction_df["DATE"].max(), freq = '7d')
```

```python
def map_week(index, row):
    if (lst_week[1] - row["DATE"]).days > 0:
        return 1
    else:
        i = 1
        while((i < len(lst_week)) & ((lst_week[i] - row["DATE"]).days < 0)):
            i += 1
            if (i  == len(lst_week) -1):
                break
        return i+1
```

```python
for index, row in transaction_df.iterrows():
    transaction_df.at[index, "Week"] = map_week(index, row)
transaction_df.sample(5)
```

```python
week_group_by = transaction_df.groupby("Week", as_index = False)
week_user_group_by = transaction_df.groupby(["Week", "User_id"], as_index = False).agg({"Order_id": "count"}).query("Order_id > 2")
week_with_users = week_group_by["User_id"].nunique().rename(columns = {"User_id": "Total Users"})
week_with_users_retention = week_user_group_by.groupby("Week", as_index = False).agg({"User_id": "count"}).rename(columns = {"User_id": "Total Users with Retention"})
weekly_rentention = week_with_users_retention.merge(week_with_users, how = "inner")
weekly_rentention["Rentention Rate"] = round((weekly_rentention["Total Users with Retention"] * 100) / weekly_rentention["Total Users"], 2)
weekly_rentention.sort_values(by = "Week", inplace = True)
```

```python
sns.lineplot(data = weekly_rentention, x = "Week", y = "Rentention Rate")
plt.title("Weekly Retention Chart")
plt.ylabel("Retention Rate (%)")
plt.show()
```

```python
## 2.2
# transaction_df["% Cashback"]
optimize_cashback_df = transaction_df.groupby(["User_id", "% Cashback"], as_index = False).agg({"GMV": "sum", "Order_id": "count"})
optimize_cashback_df.rename(columns = {"Order_id": "Total Orders"}, inplace = True)
optimize_cashback_df["Cashback"] = np.where(optimize_cashback_df["% Cashback"] != 0, "Yes", "No")
optimize_cashback_df["Retention"] = np.where(optimize_cashback_df["Total Orders"] > 1, "Yes", "No")
optimize_cashback_df.sample(5)
```

```python
cashback_retention_df = optimize_cashback_df[["Cashback", "Retention"]].value_counts().reset_index()
cashback_retention_df.rename(columns = {0: "A Number Users With Retention"}, inplace = True)
cashback_retention_df = cashback_retention_df[cashback_retention_df["Retention"] == "Yes"]
cashback_retention_df
```

```python
sns.barplot(data = cashback_retention_df, y = "A Number Users With Retention", x = "Cashback")
plt.title("Correlation of Cashback and Retention")
plt.show()
```

```python
## Check Cashback effect
cashback_gmv_df = optimize_cashback_df.groupby(["Cashback"], as_index = False).agg({"GMV": "sum"})
cashback_gmv_df
```

```python
sns.barplot(data = cashback_gmv_df, y = "GMV", x = "Cashback")
plt.title("Correlation of Cashback and GMV")
plt.show()
```

```python
diamond_user_df = transaction_df[transaction_df["Rank_name"] == "DIAMOND"]
diamond_user_df.sample(5)
```

```python
saved_diamond_uid_dict = {}
for index, row in diamond_user_df.iterrows():
    if row["User_id"] not in saved_diamond_uid_dict.keys():
        saved_diamond_uid_dict[row["User_id"]] = {row["DATE"]}
    else:
        saved_diamond_uid_dict[row["User_id"]].add(row["DATE"])  
```

```python
## Convert set to list to access by index @-@
lst_keys = list(saved_diamond_uid_dict.keys())
for i in range(len(lst_keys)):
    saved_diamond_uid_dict[lst_keys[i]] = sorted(list(saved_diamond_uid_dict.get(lst_keys[i])))
```

```python
lst_winner = []
lst_keys = list(saved_diamond_uid_dict.keys())
for i in range(len(saved_diamond_uid_dict.keys())):
    if(len(saved_diamond_uid_dict[lst_keys[i]]) > 1):
        accumulated_days = 0
        lst_days = saved_diamond_uid_dict[lst_keys[i]]
        index = len(lst_days) -1
        while(index > 0):
            accumulated_days += (lst_days[index] - lst_days[index -1]).days
            index -= 1
        if (accumulated_days >= 20):
            lst_winner.add({lst_keys[i]: accumulated_days})    
lst_winner       
```
